```python
#Calculating the Average Basepay
#we need to first fill the Na Values in the BasePay Field
#sal=sal.replace("Not Provided",np.nan)
sal

```

```python
arr=np.array(sal["BasePay"])
refreshed_arr=[]
for x in arr:
  if x == "Not Provided":
    refreshed_arr.append(np.nan)
  else:
    refreshed_arr.append(float(x))
refreshed_arr
sal.drop("BasePay",axis=1,inplace=True)
sal["BasePay"]=refreshed_arr
sal
```

```python
#getting mean of basepat
sal["BasePay"].mean()
```

```python
#Getting Highest Amount of Overtime from Dataset
arr=np.array(sal["OvertimePay"])
refreshed_arr=[]
for x in arr:
  if x == "Not Provided":
    refreshed_arr.append(np.nan)
  else:
    refreshed_arr.append(float(x))
refreshed_arr
sal.drop("OvertimePay",axis=1,inplace=True)
sal["OvertimePay"]=refreshed_arr
sal

sal["OvertimePay"].max()
```
