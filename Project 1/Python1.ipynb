```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/Data Science/14.07.2022/titles.csv")
df1=df
```

```python
df1
```

```python
df1=df[["id","title","type","release_year","age_certification","runtime","production_countries","imdb_score","imdb_votes"]]
df1=df1.sort_values(by="release_year",ascending=True)
df1
```

```python
df1.loc[:,"imdb_votes"]=df1.loc[:,"imdb_votes"].fillna(value=df1["imdb_votes"].mean())
```

```python
df1.loc[:,"imdb_score"]=df1.loc[:,"imdb_score"].fillna(value=df1["imdb_score"].std())
df1
df1.loc[:,"runtime"]=df1.loc[:,"runtime"].fillna(value=df1["runtime"].std())
df1
```

```python
df1.drop("age_certification",axis=1,inplace=True)
df1
```

```python
arr=df1["production_countries"]
arr=np.array(arr)
print(arr)
arr=list(map(lambda x:len(x),arr))
for i,x in enumerate(arr):
  if x==2:
    arr[i]=0
  else:
    arr[i]=x//6
print(arr)
df1["Shoot Locations"]=arr
df1
```

```python
df1=df1.sort_values(by="imdb_score",ascending=False).sort_values(by="imdb_votes",ascending=False).sort_values(by="Shoot Locations",ascending=False)
df1
```

```python
df1.columns
```

```python
df_g=pd.pivot_table(df1,values="runtime",index=["release_year","Shoot Locations"],columns="type",aggfunc=np.mean)
```

```python
df_g.fillna(0,inplace=True)
df_g=df_g.sort_values(by=["release_year"],ascending=False)
df_g
```
