```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Title")
ax.plot(x,y)
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([.2,.5,.2,.2])
ax1.plot(x,y,color="#ff0000")
ax2.plot(x,y,color="#ff0000")

ax1.set_xlabel("X")
ax2.set_xlabel("X")
ax1.set_ylabel("Y")
ax2.set_ylabel("Y")
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([.2,.5,.4,.4])

x=np.arange(0,101,1)
y=x**2

ax1.plot(x,y,linewidth=2)


ax2.plot(x,x*2,color="#12ff3f")
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(10,5))

x=np.arange(1,101,1)
#plotting the first graph
y=x*2
axes[0].plot(x,y,linewidth=3,linestyle="--",label="Y=2X")

#plotting the second graph
y=x**2
z=x+1
axes[0].plot(x,z,label="Y=X+1")
axes[1].plot(x,y,color="#ff0000",linewidth=4,label="y=x**2")
plt.tight_layout()
axes[0].legend()
axes[1].legend()
```
