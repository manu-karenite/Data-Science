```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
df=sns.load_dataset("titanic")
df.head()
```

```python
sns.jointplot(x="fare",y="age",data=df)
```

```python
sns.distplot(df["fare"],kde=False,color="red",bins=30)
```

```python
sns.boxplot(x="class",y="age",data=df)
```

```python
sns.swarmplot(x="class",y="age",data=df,palette="gnuplot2")
```

```python
sns.countplot(x="sex",data=df,palette="magma")
```

```python
corrdf=df.corr()
corrdf
```

```python
sns.heatmap(corrdf,cmap="cividis")
```

```python
g = sns.FacetGrid(data=df,col='sex',row="who")
g.map(plt.hist,'age')
```
